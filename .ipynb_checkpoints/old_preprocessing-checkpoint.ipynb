```python
import pandas as pd

df = pd.read_excel('Final Weighted Dataset - complete interviews (labels).xlsx')
# save to csv for future usage (it is much faster to read) I used CUDF here but pandas is also faster with csv

#Setup and point to our data file
#path = '../../data/'
#file = 'FWD_CI_Labels_all.csv'
#file = 'FWD_complete_all.csv'
#filePath = path+file

#using CUDF and the cuda/rapids code let's read the file from disk and load
#df = cu.read_csv(filePath)

df.info()

import numpy as np
# Combine 3-stage model questions into one
df3 = df.iloc[:,159:479]

df3.drop(df3.columns[[0]], axis = 1, inplace = True) # Drop index
df3.replace('I do not know', '-1', inplace=True)
df3.replace('N/A', np.NaN, inplace=True)
df3.replace('nan', np.NaN, inplace=True)
df3.replace('NaN', np.NaN, inplace=True)
```

```python
df3.to_csv("three_stage_model_questions.csv", encoding='utf-8' ,index=False)
```

```python
for columnid in range(0,79):
    df3.iloc[:,columnid] = df3.iloc[:,columnid].fillna(df3.iloc[:,columnid+80])
    df3.iloc[:,columnid] = df3.iloc[:,columnid].fillna(df3.iloc[:,columnid+160])
    df3.iloc[:,columnid] = df3.iloc[:,columnid].fillna(df3.iloc[:,columnid+240])

df4 = df3.iloc[:,0:79]
df4.to_csv("three_stage_model_questions.csv", encoding='utf-8' ,index=False)
```

```python
# Drop demographics, background variables etc...
todrop = [*range(0,18),72,74,76,85,87,89,91,93,104,106,125,*range(137,153),*range(158,479)]
#print(df.columns[todrop])

# Create df_dropped out of dropped variables demographics, background variables, etc...
df_dropped = df[df.columns[todrop]]

df5 = df.drop(df.columns[todrop], axis = 1)

df5 = df5.join(df4)

df5.drop(list(df5.filter(regex = '[OPEN]')), axis = 1, inplace = True)
df5.describe()

print("df5..")

X = df5.copy()
X.replace('I do not know', '-1', inplace=True)
#X.fillna(0, inplace=True)
X.dropna(axis=0, how='all', inplace=True)
X.dropna(axis=1, how='all', inplace=True)

# Evaluate scales for various ordinal columns
def evaluate(values):
    if values is None:
        #print('None type')
        return 0
    if values is not None and isinstance(values, str) and values.isnumeric():  # Check if value is not None, a string, and numeric
        return int(values)
    elif values == 'Significantly decrease' or values == 'Disagree' or values == 'Never' or values == 'never' or values == 'Strongly distrust' or values == '1 - very inaccurate' or values == '1 - not at all in favour' or values == '1 - very unfair' or values == '1 - not willing at all' or values == 'Fully distrust':
        return 1
    elif values == 'Decrease' or values == 'Somewhat disagree' or values == 'Rarely' or values == 'rarely' or values == 'Somewhat distrust' or values == 'Tend not to trust':
        return 2
    elif values == 'Same as today' or values == 'Neutral' or values == 'Occasionally' or values == 'Undecided':
        return 3    
    elif values == 'Increase' or values == 'Somewhat agree' or values == 'Often' or values == 'often' or values == 'Somewhat trust' or values == 'Tend to trust':
        return 4 
    elif values == 'Significantly increase' or values == 'Agree' or values == 'Always' or values == 'always' or values == 'Strongly trust' or values ==  'Fully trust' or values == '5 - very accurate':
        return 5 
    elif values == '7 - very much in favour' or values == '7 - very fair' or values == '7 - very willing':
        return 7
    elif values == 'Prefer not to say':
        return -2
    else: 
        #print(f'Value is: {values} returning 0')
        return 0

# Set scales for various ordinal columns
for column in X:
    if X[column].astype(str).str.contains('crease').any() or X[column].astype(str).str.contains('agree').any() or X[column].astype(str).str.contains('trust').any() or X[column].astype(str).str.contains('Occasionally').any() or X[column].astype(str).str.contains('Undecided').any() or X[column].astype(str).str.contains('eutral').any() or X[column].astype(str).str.contains('- very').any() or X[column].astype(str).str.startswith('1 - ').any():
        X[column] = X[column].apply(lambda x: evaluate(x)).astype(int)
        #print(X[column])
        
forbiddenChars = {' ': '_',
	';': '',
	",": "",
	"%": "perc",
	":": "",
	}
for k, v in forbiddenChars.items():
		X = X.replace(k,v, regex=True) 
		        
X.columns = X.columns.str.replace(';', '')
X.columns = X.columns.str.replace(',', '')
```
