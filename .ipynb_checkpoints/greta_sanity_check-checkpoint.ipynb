```python
res = df.isna().sum()
```

```python
miss_list=[]
for col,val in res.items():
    #if val> 5000:
    #print(col,val)
    miss_list.append([col,val])
```

```python
df1 = pd.DataFrame(miss_list, columns=["Column_Name", "Missing_Count"])
```

```python
df1.head()
```

```python
df1.sort_values(by=['Missing_Count'], ascending=True, inplace=True)
```

```python
df1.head()
```

```python
df1.to_csv("missing_count.csv", encoding='utf-8' ,index=False)
```

```python
df1.shape[0]
```

```python
mis_col = []
miss_list = []
missing_count = df1.shape[0]*0.30
for col,val in res.items():
    if val> missing_count:
        #print(col,val)
        mis_col.append(col)
        miss_list.append([col,val])
```

```python
df1 = df.drop(columns=mis_col)
```
